```python
import pytesseract
from pdf2image import convert_from_path
```

```python
pages = convert_from_path('/Users/octo/Documents/data-Sante/ce-maj.pdf', 300)  # 300 is the DPI (dots per inch)

first_page_df = None
for page_number, page in enumerate(pages):
    # Extract detailed OCR data as a dictionary
    data = pytesseract.image_to_data(page, output_type=Output.DICT)
    text = pytesseract.image_to_string(page)
    

    
    # (Optional) Print the extracted text or data
    print(f"Text from page {page_number + 1}:\n{text}\n")
    # Convert the dictionary to a DataFrame
    df = pd.DataFrame(data)
    
    # Add a column to indicate the page number (optional)
    df['page_number'] = page_number + 1
    if page_number == 0:
        first_page_df = df.copy()
    
    # Append the DataFrame to the list
    dfs.append(df)

# Combine all the DataFrames into one (if there are multiple pages)
final_df = pd.concat(dfs, ignore_index=True)
```
